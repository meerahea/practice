Que2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
#Note: Also add title, xlabel, and ylabel to the plot.
```

Ans:

A scatter plot is a type of data visualization that displays individual data points as dots on a 2D plane. It is particularly useful for visualizing the relationship between two variables and identifying any patterns, trends, or clusters within the data.

```python
# Create a scatter plot
plt.figure(figsize=(8, 6))  # Set the figure size

plt.scatter(x, y, c='blue', marker='o', label='Data Points')  # Scatter plot with blue circles
plt.title('Scatter Plot Example')  # Title for the plot
plt.xlabel('X-axis')  # Label for the x-axis
plt.ylabel('Y-axis')  # Label for the y-axis
plt.legend()  # Display legend

plt.grid(True)  # Add gridlines

plt.show()  # Display the plot
```

Que3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:

```python
import matplotlib.pyplot as plt
import numpy as np
# Data for the line plots
x = np.array([0, 1, 2, 3, 4, 5])

# Create subplots
plt.figure(figsize=(12, 8))

# Line plot 1
plt.subplot(2, 2, 1)
y1 = np.array([0, 100, 200, 300, 400, 500])
plt.plot(x, y1, marker='o')
plt.title('Line Plot 1')

# Line plot 2
plt.subplot(2, 2, 2)
y2 = np.array([50, 20, 40, 20, 60, 70])
plt.plot(x, y2, marker='o', color='orange')
plt.title('Line Plot 2')

# Line plot 3
plt.subplot(2, 2, 3)
y3 = np.array([10, 20, 30, 40, 50, 60])
plt.plot(x, y3, marker='o', color='green')
plt.title('Line Plot 3')

# Line plot 4
plt.subplot(2, 2, 4)
y4 = np.array([200, 350, 250, 550, 450, 150])
plt.plot(x, y4, marker='o', color='red')
plt.title('Line Plot 4')

# Adjust layout
plt.tight_layout()

# Display the plots
plt.show()
```

Que4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.


Ans:

A bar plot, also known as a bar chart, is a type of data visualization that uses rectangular bars to represent data values. It's commonly used to compare and display the quantitative values of different categories. The length of each bar corresponds to the value of the data it represents, and the bars are typically arranged along one axis (vertical or horizontal), with the other axis representing the categories.
Bar plots are used for various purposes, including:
Comparisons,Trends,Ranking,Highlighting Differences,Categorical Data


```python
import matplotlib.pyplot as plt
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
# Creating a vertical bar plot
plt.figure(figsize=(10, 6))
plt.bar(company, profit, color='blue')
plt.xlabel('Company')
plt.ylabel('Profit')
plt.title('Company Profits')
plt.show()
```

```python
# Creating a horizontal bar plot
plt.figure(figsize=(10, 6))
plt.barh(company, profit, color='green')
plt.xlabel('Profit')
plt.ylabel('Company')
plt.title('Company Profits')
plt.show()
```

Que5: What is a box plot? Why is it used? Using the following data plot a box plot.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

Ans:

A box plot, also known as a whisker plot, is a graphical representation of data that displays a summary of the distribution. It provides a visual way to understand the central tendency, dispersion, and potential outliers in a dataset.

Box plots are particularly useful for Identifying Central Tendency,Understanding Spread, Spotting Outliers, Comparing Distributions:

```python
import matplotlib.pyplot as plt
import numpy as np

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Creating the box plot
data_to_plot = [box1, box2]
labels = ['Box 1', 'Box 2']

plt.boxplot(data_to_plot, labels=labels)
plt.title('Box Plot')
plt.ylabel('Value')
plt.xlabel('Box')
plt.show()
```
